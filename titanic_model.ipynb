# Splitting the data into training data & Test data


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

# **Accuracy** Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, mean_squared_error
import numpy as np

# Predictions
y_pred = model.predict(X_test)             # class predictions (0/1)
y_proba = model.predict_proba(X_test)[:,1] # probability of survival

# Standard classification metrics
print("Accuracy:", accuracy_score(Y_test, y_pred))
print("F1 Score:", f1_score(Y_test, y_pred))
print("ROC-AUC:", roc_auc_score(Y_test, y_proba))

# RMSE (using probabilities vs. actual labels)
rmse = np.sqrt(mean_squared_error(Y_test, y_proba))
print("RMSE (probabilities vs. actual):", rmse)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, mean_squared_error
import numpy as np

# Initialize Decision Tree
dt = DecisionTreeClassifier(random_state=2)

# Train
dt.fit(X_train, Y_train)

# Predictions
y_pred = dt.predict(X_test)               # class predictions (0/1)
y_proba = dt.predict_proba(X_test)[:, 1]  # survival probabilities

# Metrics
print("Accuracy:", accuracy_score(Y_test, y_pred))
print("F1 Score:", f1_score(Y_test, y_pred))
print("ROC-AUC:", roc_auc_score(Y_test, y_proba))

# RMSE (probabilities vs. actual survival)
rmse = np.sqrt(mean_squared_error(Y_test, y_proba))
print("RMSE:", rmse)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, mean_squared_error
import numpy as np

# Initialize Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=2)

# Train
rf.fit(X_train, Y_train)

# Predictions
y_pred = rf.predict(X_test)               # class predictions (0/1)
y_proba = rf.predict_proba(X_test)[:, 1]  # survival probabilities

# Metrics
print("Accuracy:", accuracy_score(Y_test, y_pred))
print("F1 Score:", f1_score(Y_test, y_pred))
print("ROC-AUC:", roc_auc_score(Y_test, y_proba))

# RMSE (probabilities vs. actual survival)
rmse = np.sqrt(mean_squared_error(Y_test, y_proba))
print("RMSE:", rmse)
```
